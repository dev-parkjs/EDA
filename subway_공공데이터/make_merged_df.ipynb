```python
pd.set_option('display.max_columns', None)
dfs = [df19, df20, df21, df23, df24, df23_24, df22, df23_1031, df23_0630]

def standardize_columns(df):
    df = df.copy()
    if '수송일자' in df.columns:
        df.rename(columns={'수송일자': '날짜'}, inplace=True)
    
    if '고유역번호(외부역코드)' in df.columns:
        df.rename(columns={'고유역번호(외부역코드)': '역번호'}, inplace=True)
    
    if '승하차구분' in df.columns:
        df.rename(columns={'승하차구분': '구분'}, inplace=True)
    
    if '06:00 이전' in df.columns:
        df.rename(columns={'06:00 이전': '06시 이전'}, inplace=True)
    if '06시이전' in df.columns:
        df.rename(columns={'06시이전': '06시 이전'}, inplace=True)
    
    rename_dict = {}
    for col in df.columns:
        if " ~ " in col:
            new_col = col.replace(" ", "").replace("~", "-")
            rename_dict[col] = new_col  
        elif ":00" in col and "~" in col:
            new_col = col.replace(":00", "").replace(" ", "").replace("~", "-")
            rename_dict[col] = new_col
        elif "시-" in col:
            new_col = col.replace("시", "")
            rename_dict[col] = new_col
        elif "시간대" in col:
            new_col = col.replace("시간대", "").strip()
            rename_dict[col] = new_col

    df.rename(columns=rename_dict, inplace=True)
    

    if '23시이후' in df.columns:
        df.rename(columns={'23시이후': '23-24'}, inplace=True)
    if '23시 이후' in df.columns:
        df.rename(columns={'23시 이후': '23-24'}, inplace=True)
    if '24:00 이후' in df.columns:
        df.rename(columns={'24:00 이후': '24시 이후'}, inplace=True)
    if '24시이후' in df.columns:
        df.rename(columns={'24시이후': '24시 이후'}, inplace=True)
    
    
    if '호선' in df.columns:
        df['호선'] = df['호선'].astype(str).apply(lambda x: x if x.endswith('호선') else x + '호선')
    
    base_cols = ['날짜', '호선', '역번호', '역명', '구분']
    time_cols = ['06시 이전', '06-07', '07-08', '08-09', '09-10', '10-11',
                 '11-12', '12-13', '13-14', '14-15', '15-16', '16-17',
                 '17-18', '18-19', '19-20', '20-21', '21-22', '22-23', '23-24', '24시 이후']
    
    if '합 계' in df.columns:
        df.rename(columns={'합 계': '합계'}, inplace=True)
    if '합계' not in df.columns:
        df['합계'] = pd.NA

    if '24시 이후' not in df.columns:
        df['24시 이후'] = pd.NA
        
    final_order = base_cols + [tc for tc in time_cols if tc in df.columns] + ['합계']
    df = df[[col for col in final_order if col in df.columns]]
    return df

dfs_std = [standardize_columns(df) for df in dfs]

merged_df = pd.concat(dfs_std, ignore_index=True)

merged_df['날짜'] = pd.to_datetime(merged_df['날짜'], errors='coerce')

merged_df = merged_df.sort_values(by='날짜')

time_cols = ['06시 이전', '06-07', '07-08', '08-09', '09-10', '10-11',
             '11-12', '12-13', '13-14', '14-15', '15-16', '16-17',
             '17-18', '18-19', '19-20', '20-21', '21-22', '22-23', '23-24', '24시 이후']

merged_df['합계'] = merged_df[time_cols].apply(pd.to_numeric, errors='coerce').fillna(0).sum(axis=1)

merged_df
```
